```python
#function to read file and put into dataframe
def read_data(filename,col1="Pop",col2="Pro"):
    data = pd.read_table(filename, sep=",", header = None)
    data.index = list(range(1, data.shape[0]+1))
    data.columns = list([col1,col2])
    return data
```

```python
data = read_data('ex1data1.txt')
data.head()

```

```python
# function to plot scatterplot for preliminary data view
def plot_data(data,col1="Pop",col2="Pro"):
    fig, ax = plt.subplots()
    ax.scatter(data[col1], data[col2])
    plt.xlabel(col1)
    plt.ylabel(col2)
    
    x = ax.get_xgridlines()
    x1 = x[1]
    x1.set_color('black')

    y = ax.get_ygridlines()
    y1 = y[1]
    y1.set_color('black')


    plt.xlim(-5, 30) 
    plt.ylim(-5, 30)
    #return data
```

```python
plot_data(data)
```

```python
#function that takes in data dataframe and return dataframes of examples(X), parameters(theta)
#and indicators(Y)
def make_X_theta_Y(data):
    
    theta = np.zeros(shape=data.shape[1])
    theta = pd.DataFrame(theta)
    #theta.index = ['theta_0','theta_1']
    theta.columns = ['theta']
    
    Y = data.loc[:,["Pro"]]
    
    X_Rest = data.loc[:,~data.columns.isin(['Pro'])]
    
    X_0 = np.ones(shape=data.shape[0])
    X_0 = pd.DataFrame(X_0)
    X_0.index = list(range(1, X_0.shape[0]+1))
    
    X = pd.concat([X_0,X_Rest],axis=1)
    X = X.rename(columns = {0:'X_0'})
    
    return theta,X,Y
```

```python
theta,X,Y = make_X_theta_Y(data)
```

```python
#function that takes in parameters(theta) and examples (X) and returns hypothesis (h)
def hypothesis(theta,X):
    ind = list(range(1, X.T.shape[1]+1))
    h = np.dot(X,theta)
    h = pd.DataFrame(h,index=ind,columns=[1]) 
    return h
```

```python
h = hypothesis(theta,X)
#h.head()
```

```python
#function that calculates the cost (J) from the hypothesis(h) and indicators(Y)
def error_cost(h,Y):
    
    a = h.iloc[:,0].subtract(Y.iloc[:,0])
    b = a.iloc[:,].multiply(a.iloc[:,])

    cost = pd.DataFrame(b).sum()
    cost = cost/(2*Y.shape[0])
    return cost
```

```python
cost = error_cost(h,Y)
cost
```

```python
#function that updates theta using batch gradient descent
def gradient_descent(theta,X,Y,alpha):
       
    h = hypothesis(theta,X)
    a = h.iloc[:,0].subtract(Y.iloc[:,0]) 
    
    g = pd.DataFrame([0,0],index=theta.index,columns=theta.columns) 
    
    for i in range(0,len(g.index)):
        
        holder = a.iloc[:,].multiply(X.iloc[:,i]) 
        holder = pd.DataFrame(holder).sum()
        holder = (holder*alpha)/(Y.shape[0])
        
        g.ix[[i],['theta']] = holder[0]
        
    new_theta = theta.iloc[:,0].subtract(g.iloc[:,0])

    return pd.DataFrame(new_theta,index=theta.index,columns=theta.columns) 
    
```

```python
#set number of iterations of gradient descent to run and the value of alpha 
iterations = 1500
alpha = 0.01
```

```python
#run gradient descent and save the changing values of the parameters(to_plot_0,to_plot_1)
#and cost (to_plot_j)
to_plot_0 = []
to_plot_1 = []
to_plot_j = []

updated_theta = theta.copy()
to_plot_0.append(updated_theta.iloc[0,:])
to_plot_1.append(updated_theta.iloc[1,:])
to_plot_j.append(error_cost(hypothesis(updated_theta,X),Y))

for j in range(0,iterations):
    updated_theta = gradient_descent(updated_theta,X,Y,alpha)
    to_plot_0.append(updated_theta.iloc[0,:])
    to_plot_1.append(updated_theta.iloc[1,:])
    to_plot_j.append(error_cost(hypothesis(updated_theta,X),Y))
```

```python
#convert lists containing values of parameters and costs into dataframes and slice for plotting
to_plot_0 = pd.DataFrame(to_plot_0)
to_plot_1 = pd.DataFrame(to_plot_1)
to_plot_j = pd.DataFrame(to_plot_j)

t0_full_df = to_plot_0.copy()
t1_full_df = to_plot_1.copy()
j_full_df = to_plot_j.copy()

t0_sliced_df = to_plot_0[0::7]
t1_sliced_df = to_plot_1[0::7]
j_sliced_df = to_plot_j[0::7]
```

```python
#function to calculate error cost from individual theta
def error_cost_theta(tt0,tt1):
    tta = pd.DataFrame(list([tt0,tt1]))
    h = hypothesis(tta,X)
    cost = error_cost(h,Y)
    return cost
    
    
```
